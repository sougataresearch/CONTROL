# Notebook 1 — Motor Bring-Up (interactive, confirmation-gated)
Full startup sequence in your exact order, **one motor at a time**, with
settling gaps between motors and a **y/N permission prompt before every stage**:

detect → connect → initialize → enable → home → move to optical zero

> **Mode is selected FIRST** — it decides how many motors are used
> (3×3 → 2 motors, 4×4 → 4 motors).

```
# Cell 1 -- select the measurement mode FIRST (this decides the motor count)
import sys; sys.path.insert(0, ".")                  # make mmie importable
from mmie.motors import MotorBank                    # the grouped controller
print("Modes:  1 = 3x3 (2 polarizers)   2 = 4x4 discrete (4 motors)   3 = 4x4 continuous (4 motors)")
choice = input("Select mode [1/2/3]: ").strip()      # ask in the notebook prompt
mode = {"1": "3x3", "2": "4x4_discrete", "3": "4x4_continuous"}[choice]  # map to mode name
bank = MotorBank(mode)                               # builds ONLY the motors this mode needs
```

```
# Cell 2 -- run the whole gated bring-up sequence
# Each stage asks for your y/N permission, runs motor-by-motor with settling
# gaps, and prints a confirmation per motor and per stage.
bank.full_bringup()                                  # detect->connect->init->enable->home->zero
```

```
# Cell 3 -- OPTIONAL manual sanity move: send ONE motor to ONE optical angle
name = input("Motor to test (e.g. PSG_POL), or Enter to skip: ").strip()
if name:                                             # only if a name was typed
    ang = float(input("Optical angle in degrees: ")) # e.g. 30
    bank.motors[name].move_to_optical(ang)           # applies zero offset + wrap, then moves
    bank.motors[name].go_to_optical_zero()           # return to optical zero afterwards
```

```
# Cell 4 -- run this cell ONLY when you are completely done with the motors
bank.shutdown_all()                                  # stop polling + disconnect every motor
```
