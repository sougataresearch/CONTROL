# Notebook 2 — State Generation (dry run, **no hardware needed**)
Test the angle rules on any computer:
- `"360/10"` → 36 positions (360° ≡ 0°, so 360 is excluded)
- `"0,30,60,90"` → exactly those angles
- PSG and PSA lists may be **different**; grid = full cross product
- All inputs are **optical** angles; motor angles = (zero_offset + optical) mod 360

```
# Cell 1 -- parse any spec and see the resulting optical angle list
import sys; sys.path.insert(0, ".")
from mmie.angles import parse_angle_spec, build_combinations, preview_combinations
spec_psg = input("PSG angle spec (e.g. 360/10  or  0,30,60,90): ").strip()   # your input style
spec_psa = input("PSA angle spec (may be different):            ").strip()
psg = parse_angle_spec(spec_psg)                     # -> list of optical angles
psa = parse_angle_spec(spec_psa)
print(f"PSG angles ({len(psg)}): {psg}")             # print full lists so you can verify
print(f"PSA angles ({len(psa)}): {psa}")
```

```
# Cell 2 -- build and preview the measurement grid (what WILL be measured)
combos = build_combinations(psg, psa)                # every PSG paired with every PSA
preview_combinations(combos, n_show=10)              # first 10 + total count
```

```
# Cell 3 -- verify optical -> motor conversion for each configured motor
from mmie import config
from mmie.angles import optical_to_motor
test_angle = float(input("Test OPTICAL angle (e.g. 320): "))    # try 320 to see the wrap
for name, m in config.MOTORS.items():               # apply each motor's own offset
    motor_ang = optical_to_motor(test_angle, m["zero_offset"])  # (offset+optical) % 360
    print(f"  {name:8s}: optical {test_angle} -> motor {motor_ang}  (offset {m['zero_offset']})")
```
